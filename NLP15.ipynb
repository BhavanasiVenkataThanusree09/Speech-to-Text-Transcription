```python
import speech_recognition as sr
from pydub import AudioSegment
import os
def convert_to_wav(file_path):
    if not file_path.lower().endswith(".wav"):
        wav_file = "temp.wav"
        AudioSegment.from_file(file_path).export(wav_file, format="wav")
        return wav_file
    return file_path
r = sr.Recognizer()
def mic_speech_to_text(language='en-IN'):
    try:
        with sr.Microphone() as source:
            print(f"Speak something ({language}): ")
            r.adjust_for_ambient_noise(source)
            audio = r.listen(source)
        text = r.recognize_google(audio, language=language)
        print("You said:", text)
    except sr.UnknownValueError:
        print("Sorry, could not understand the audio.")
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))
def file_speech_to_text(file_path, language='en-IN'):
    try:
        wav_file = convert_to_wav(file_path)
        with sr.AudioFile(wav_file) as source:
            audio = r.record(source)
        text = r.recognize_google(audio, language=language)
        print("You said:", text)
        if wav_file == "temp.wav":  # delete temp file if created
            os.remove(wav_file)
    except FileNotFoundError:
        print("File not found:", file_path)
    except sr.UnknownValueError:
        print("Sorry, could not understand the audio.")
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))
    except Exception as e:
        print("Error:", e)
while True:
    print("\nOptions:")
    print("1. Speak using Microphone")
    print("2. Recognize from Audio File")
    print("3. Exit")
    choice = input("Enter choice (1/2/3): ").strip()

    if choice == "1":
        lang = input("Enter language code ('en-IN' for English, 'te-IN' for Telugu): ").strip() or 'en-IN'
        mic_speech_to_text(language=lang)
    elif choice == "2":
        file_path = input("Enter audio file path: ").strip()
        lang = input("Enter language code ('en-IN' for English, 'te-IN' for Telugu): ").strip() or 'en-IN'
        file_speech_to_text(file_path, language=lang)[-]
    elif choice == "3":
        print("Exiting...")
        break
    else:
        print("Invalid choice. Try again.")
```
